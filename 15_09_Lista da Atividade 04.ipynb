# **1 - Questão de Velocidade Média**

Um carro percorreu uma certa distância em um período de tempo. Você precisa criar um programa Python que calcule a velocidade média desse carro. O programa deve pedir ao usuário que insira a distância percorrida em quilômetros e o tempo levado em horas. Em seguida, ele deve calcular a velocidade média em quilômetros por hora e exibir o resultado.

**Aqui estão os passos para a questão:**

Solicitar ao usuário que insira a distância percorrida em quilômetros (um número decimal).

Solicitar ao usuário que insira o tempo levado em horas (um número decimal).

Calcular a velocidade média usando a fórmula:

Velocidade Média = Distância / Tempo

Exibir a velocidade média calculada no formato "A velocidade média do carro foi de X km/h", onde X é o resultado calculado.



```python
distancia = float(input('Informe a distância percorrida em Km: '))
tempo = float(input('Informe o tempo gasto em horas: '))

velocidade_med = distancia / tempo

print(f'A velocidade médoa do carro foi {velocidade_med} Km/h')
```

# **2 - Questão de Geometria: Cálculo de Área de um Triângulo**

Você precisa criar um programa Python que calcule a área de um triângulo com base nas medidas de sua base e altura. O programa deve pedir ao usuário que insira a base e a altura do triângulo e, em seguida, calcular a área e exibir o resultado.

**Aqui estão os detalhes:**

Solicitar ao usuário que insira a medida da base do triângulo (em metros).

Solicitar ao usuário que insira a medida da altura do triângulo (em metros).

Calcular a área do triângulo usando a fórmula:

Área = (Base * Altura) / 2

Exibir a área calculada no formato "A área do triângulo é X metros quadrados", onde X é o resultado calculado.


```python
base = float(input('Informe a base do triângulo em metros: '))
altura = float(input('Informe a altura do triângulo em metros: '))

area = (base * altura) / 2

print(f'A área do triângulo é {area} metros quadrados')
```

# **3 - Estudo de Caso: Cálculo de Juros Simples**

Você é um consultor financeiro e está ajudando um cliente a calcular os juros simples para um investimento. O cliente possui um portfólio de 03 investimentos em que cada investimento tem uma taxa de juros diferente e um período de investimento diferente. Você precisa calcular o montante total após o fim do período de investimento para cada investimento e, em seguida, somar todos os montantes para determinar o valor total.

**Aqui estão os detalhes:**

O cliente possui três investimentos:

Investimento 1: Principal de R$ 10.000, taxa de juros de 5% ao ano e período de investimento de 3 anos.

Investimento 2: Principal de R$ 15.000, taxa de juros de 7.5% ao ano e período de investimento de 2 anos.

Investimento 3: Principal de R$ 20.000, taxa de juros de 4% ao ano e período de investimento de 5 anos.

Crie um programa Python que calcule o montante total após o período de investimento para cada investimento usando a fórmula de juros simples:

Em seguida, some todos os montantes para determinar o valor total do portfólio do cliente.


```python
investimento1 = 10000 + (10000 * 0.05 * 3)
investimento2 = 15000 + (15000 * 0.075 * 2)
investimento3 = 20000 + (20000 * 0.04 * 5)



montante_total = investimento1 + investimento2 + investimento3

print(f'Montante 1: R$ {investimento1:.2f}')
print(f'Montante 2: R$ {investimento2:.2f}')
print(f'Montante 3: R$ {investimento3:.2f}')
print(f'Montante toal: R$ {montante_total:.2f}')
```

# **4 - Estudo de Caso: Rastreando Investimentos com Juros Simples**

Você está criando um programa para ajudar um investidor a rastrear seu investimento ao longo do tempo com base em juros simples. O investidor deseja saber quanto seu investimentos valerá em um número específico de anos. Para isso, você deve criar um programa Python que permita ao investidor adicionar investimentos, calcular seus valores futuros e fornecer uma previsão de quanto o portfólio valerá em um período especificado.

**Aqui estão os detalhes:**

O programa deve permitir que o investidor adicione investimentos, especificando o principal, a taxa de juros anual e o período de investimento.

O programa deve calcular o valor futuro de cada investimento usando a fórmula de juros simples:

Montante = Principal + (Principal * Taxa de Juros * Tempo)


O investidor também deve ser capaz de especificar o número de anos no futuro para calcular o valor total do investimento.

Finalmente, o programa deve imprimir o valor total do investimento após o período especificado.


```python
capital = float(input('Informe o capital aplicado: R$ '))
taxa = float(input('Informe a taxa de juros contratada: '))
tempo = float(input('Informe por quantos anos o capital será aplicado: '))

montante = capital + (capital * (taxa / 100)  * tempo)

print(f'Com um capial de  R$ {capital:.2f} aplicado por {tempo:.0f} ano(s) a uma taxa de {taxa:.2f}%\nO investidor obterá um montante de R$ {montante:.2f}')
```

**5 - Questão de Matemática Financeira: Juros Compostos e Porcentagem**

Um investidor está considerando colocar dinheiro em uma conta de poupança com juros compostos. Ele deseja entender quanto dinheiro ele terá após um determinado número de anos, levando em consideração a taxa de juros anual da conta. Você deve criar um programa Python que calcule o montante total após um período de tempo especificado e, em seguida, determine quanto esse montante representa em relação ao valor inicial. O programa deve pedir ao usuário que insira o valor inicial, a taxa de juros anual e o número de anos. Em seguida, ele deve calcular o montante final e a porcentagem de aumento em relação ao valor inicial.

**Aqui estão os detalhes:**

Solicitar ao usuário que insira o valor inicial em reais.

Solicitar ao usuário que insira a taxa de juros anual (em decimal, por exemplo, 5% é 0.05).

Solicitar ao usuário que insira o número de anos para o investimento.

Calcular o montante final usando a fórmula de juros compostos:

Montante = Valor Inicial * (1 + Taxa de Juros) ^ Anos

Calcular a porcentagem de aumento em relação ao valor inicial usando a fórmula:

Porcentagem de Aumento = ((Montante - Valor Inicial) / Valor Inicial) * 100

Exibir o montante final e a porcentagem de aumento no formato "O montante final após X anos é de R$ Y, representando um aumento de Z% em relação ao valor inicial".


```python
capital = float(input('Informe o capital aplicado: R$ '))
taxa = float(input('Informe a taxa de juros contratada em decimal: '))
tempo = float(input('Informe por quantos anos o capital será aplicado: '))

montante = capital * (1 + taxa) ** tempo
variacao = ((montante - capital) / capital) * 100


print(f'Após {tempo} anos a aplicação de R$ {capital:.2f} a uma taxa de {(taxa * 100):.2f}% resultará num montante de R$ {montante:.2f}, representando um aumento de {variacao:.2f}%')
```

Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o Imposto de Renda, 8% para o INSS e 5% para o sindicato, faça um programa que nos dê: salário bruto. quanto pagou ao INSS. quanto pagou ao sindicato. o salário líquido. calcule os descontos e o salário líquido, conforme a tabela abaixo:

Salário Bruto : R −IR(11
INSS (8%) : R −Sindicato(5  = Salário Liquido : R$

```python
valor_hora = float(input('Informe o valor da hora trabalhada: R$ '))
horas_trabalho = int(input('Informe o número de horas trabalhadas: '))

salario_bruto = valor_hora * horas_trabalho
ir = 0.11
inss = 0.08
sindicato = 0.05
desconto_ir = salario_bruto * ir
desconto_inss = salario_bruto * inss
desconto_sindical = salario_bruto * sindicato
salario_liquido = salario_bruto - desconto_ir - desconto_inss - desconto_sindical

print(f'Salário Bruto: \t\t\tR$ {salario_bruto:.2f}')
print(f'Imposto de Renda:\t\tR$ ({desconto_ir:.2f})')
print(f'INSS:\t\t\t\tR$ ({desconto_inss:.2f})')
print(f'Desconto sindical:\t\tR$ ({desconto_sindical:.2f})')
print(f'Salário Líquido: \t\tR$ {salario_liquido:.2f}')
```
