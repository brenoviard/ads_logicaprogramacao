# Questão 01 - Calculadora de Média

```python
nota1 = float(input('Informe a primeira nota: '))
nota2 = float(input('Informe a segunda nota: '))
nota3 = float(input('Informe a terceira nota: '))

media = (nota1 + nota2 + nota3)/3

media = round(media,2)

print(f'Média artimética: {media}')
```

# Questão 02 - Média Ponderada


```python
nota1 = float(input('Informe a primeira nota: '))
nota2 = float(input('Informe a segunda nota: '))
nota3 = float(input('Informe a terceira nota: '))

media_arit = (nota1 + nota2 + nota3)/3
media_pond = ((nota1 * 1) + (nota2 * 2)+ (nota3 * 3) ) / 6

media_arit = round(media_arit, 2)
media_pond = round(media_pond, 2)

print('RESULTADO')
print(f'Média Aritmética: {media_arit}')
print(f'Média Ponderada: {media_pond}')
```

# Questão 03 - Conversor de Moedas

```python
valor_dolar = float(input('Informe o valor em dolares: '))

cotac_euro = 0.93

valor_euro = valor_dolar * cotac_euro

print(f'{valor_dolar} doláres equivale a {valor_euro} euros')
```

# Questão 04 - Calculadora de Gorjeta

```python
valor_conta = float(input('Informe o valor da conta: '))
taxa_gorjeta = float(input('Informe a porcentagem da gorjeta: '))

gorjeta = round((taxa_gorjeta/100) * valor_conta,2)
valor_total = valor_conta + gorjeta

print(f'Conta: {valor_conta}')
print(f'Gorjeta: {gorjeta}')
print(f'Total: {valor_total}')
```
