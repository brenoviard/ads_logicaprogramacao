Questão 1: Cálculo de Juros Compostos:

Elabore um fluxograma que calcule o montante final de um investimento com base no capital inicial, taxa de juros anual e número de anos. O programa deve utilizar a fórmula de juros compostos:

M = P  (1+ r/100)n

Onde:

M é o montante final.

P é o capital inicial.

r é a taxa de juros anual.

n é o número de anos.

O programa deve solicitar ao usuário os valores de P, r e n e exibir o montante final

```python
p = float(input('Informe o Capital Inicial: '))
r = float(input('Informe a taxa de juros anual: '))
n = float(input('Informe o período do investimento: '))

m = p*(1+(r/100))**n

print(m)
```

Questão 2: Calculadora de Expressões Matemáticas:

Elabore um fluxograma que funcione como uma calculadora de expressões matemáticas simples. O programa deve permitir ao usuário inserir uma expressão no formato "número operador número" (por exemplo, "5 + 3" ou "7 * 2"). O programa deve calcular e exibir o resultado da expressão.

```python
print('Calculadora')
print('Informe os valores ')
val_1 = float(input('Primeiro valor: '))
val_2 = float(input('Segundo valor: '))
print("Tipos de Operação:\n'+' soma\n'-' subtração\n'*' multiplicação\n'/' divisão")
operacao = str(input('Informe o tipo de operação: '))

if operacao =='+':
  resultado = val_1 + val_2
  print(f'Resultado: {resultado}')
elif(operacao =='-'):
  resultado = val_1 - val_2
  print(f'Resultado: {resultado}')
elif(operacao =='*'):
  resultado = val_1 * val_2
  print(f'Resultado: {resultado}')
elif(operacao =='/'):
  if(val_2 == 0):
    resultado ='Operação inválida dentro do conjunto dos números reais'
    print(f'Resultado: {resultado}')
  else:
    resultado = val_1 / val_2
    print(f'Resultado: {resultado}')
else:
  print('Operação inválida')

```

Questão 4: Sistema de Controle de Acesso:

Você deve criar um fluxograma para um sistema de controle de acesso a um edifício. O programa deve solicitar ao usuário uma identificação (cartão ou código). O sistema verificará se o código ou cartão é válido. Se for válido, o programa permitirá o acesso. Se não for válido, o programa verificará se o usuário excedeu o número máximo de tentativas 3X. Se sim, o acesso será negado temporariamente. Caso contrário, o programa permitirá que o usuário tente novamente.

```python
print("Insira o código de acesso: ")

cont = 0
while cont < 3:
  codigo = str(input('Código: '))
  if codigo == '9090':
    print('Seja bem-vindo')
    break
  else:
    if cont == 2:
      print('Usuário bloqueado, tenten novamente mais tarde')
      break
    else:
      print('Senha incorreta, tente novamente.')
      cont += 1
```

Questão 5: Sistema de Monitoramento de Saúde:

Desenvolva um fluxograma para um sistema de monitoramento de saúde. O programa deve coletar informações sobre a saúde do usuário, incluindo pressão arterial, frequência cardíaca e temperatura. Com base nas informações coletadas, o programa deverá determinar se o usuário está em bom estado de saúde ou se há indícios de preocupação, se dois dos três indícios estiverem fora do padrão o programa fornecerá recomendações apropriadas.

```python
#frequencia cardiaca padrao entren 60 e 100 bpm
#temperadura media 37º
#pressao arterial 93,3

print('Diagnóstico do Paciente: ')
freq_card = float(input('Informe a Frequência Cardíaca: '))
temp = float(input('Informe a Temperatura: '))
press_art = float(input('Informe a Pressão Arterial: '))

if freq_card >= 60 and freq_card <= 100:
  status_freq = 'OK'
else:
  status_freq = 'Alerta'
if temp >= 35 and temp <=37:
  status_temp = 'OK'
else:
  status_temp = 'Alerta'
if press_art >= 93.3 and press_art <= 113.3:
  status_press = 'OK'
else:
  status_press = 'Alerta'

if (status_freq == 'OK'):
  if(status_press == 'OK' or status_temp == 'OK'):
    print("Paciente apresenta indíces dentro do padrão.")
  else:
    print("Paciente em estado de alerta. Dirija-se a um ´médico.")
else:
  if(status_press == 'OK' and status_temp == 'OK'):
    print("Paciente apresenta indíces dentro do padrão.")
  else:
    print("Paciente em estado de alerta. Dirija-se a um ´médico.")

```

Questão 6: Sistema de Reservas de Passagens:

Crie um fluxograma para um sistema de reservas de passagens aéreas. O programa deve permitir ao usuário escolher um destino, data e classe de voo. Em seguida, o programa verificará a disponibilidade de assentos na classe selecionada. Se houver assentos disponíveis, o programa confirmará a reserva. Caso contrário, o programa oferecerá opções alternativas ou pedirá ao usuário para escolher outra classe.


```python
data = str(input('data disponível? True or False '))
destino = str(input('destino disponível? True or False '))
classe = str(input('Classe disponivel? True or False '))

if ((classe == 'true') and (classe == destino) and (classe == data)):
  poltrona = str(input('poltorna disponível? True or False'))
  if poltrona == 'true':
    print('Reserva realizada!')
  else:
    poltrona = bool(input('Poltrona ocupada. Selecione outra Poltrona'))
    if poltrona == 'true': print ('Reserva realizada!')
    else: print('Poltrona ocupada. Reinicie o procedimento de reserva.')
else: print('selecione uma opção válida!')

```
